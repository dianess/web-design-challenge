# Convert Dow Jones Data to html

```python
# Dependencies
import csv
import pandas as pd
```

```python
# Read the CSV file and create a dataframe
dow_jones_df= pd.read_csv('dow_jones2.csv')
dow_jones_df.head()
```

```python
# Invert the data in the dataframe so it's chronological in order
inverted_dow_jones = dow_jones_df.iloc[::-1]
inverted_dow_jones.head()
```

```python
# Remove the unnecessary index
inverted_dow_jones_no_index=inverted_dow_jones.set_index('timestamp')
inverted_dow_jones_no_index.head()
```

```python
# Convert the pandas dataframe to html and export as "dow_jones_html_data.html"
inverted_dow_jones_no_index.to_html('dow_jones_html_data.html')
```
